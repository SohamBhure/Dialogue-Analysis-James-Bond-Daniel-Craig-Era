```python
def tokenize_and_clean(PATH):
  string = open(PATH, encoding="latin-1").read()
  string2 = nltk.word_tokenize(string)
  new_str = re.sub('[^a-zA-Z0-9\n\.]', ' ', string)
  tokens1 = nltk.word_tokenize(new_str)
  #print(tokens1)

  tokens = []
  for i in tokens1:
    if i[0]=='.' or i[0]==',' or i[0]=='-' or i[0]=='\'':
      continue
    tokens.append(i)

  return tokens, string2
```

```python
def lower_case(tokens):
  token_lower = []
  for word in tokens:
    token_lower.append(word.lower())
  
  return token_lower
```

```python
def word_cloud(token):
  remove_these = set(stopwords.words('english') + ['l'])
  filtered_text = [w for w in token if not w in remove_these]
  unique_string=(" ").join(filtered_text)
  
  wordcloud = WordCloud(width = 1000, height = 500, background_color ='white').generate(unique_string)
  plt.figure(figsize=(15,8), facecolor = None)
  plt.imshow(wordcloud)
  plt.axis("off")
  #plt.savefig("your_file_name"+".png", bbox_inches='tight')
  plt.show()
  plt.close()
```

```python
def top_words(token, title_movie):
  remove_these = set(stopwords.words('english') + ['l'])
  filtered_text = [w for w in token if not w in remove_these]
  fdist_filtered = FreqDist(filtered_text)

  plt.figure(figsize=(10,5))
  display(fdist_filtered.plot(20,title=title_movie))
```

Finding the common words between five movies

```python
def clean_and_dict(token):
  remove_these = set(stopwords.words('english') + ['l'])
  filtered_text = [w for w in token if not w in remove_these]
  dict_token = {i:filtered_text.count(i) for i in set(filtered_text)}
  return dict_token
```

```python
dict_token1 = clean_and_dict(lower_token1)
dict_token2 = clean_and_dict(lower_token2)
dict_token3 = clean_and_dict(lower_token3)
dict_token4 = clean_and_dict(lower_token4)
dict_token5 = clean_and_dict(lower_token5)
```

```python
dict_token5['bond']
```

```python
dict_token1_sorted = {k: v for k, v in sorted(dict_token1.items(), key=lambda item: item[1])}
dict_token2_sorted = {k: v for k, v in sorted(dict_token2.items(), key=lambda item: item[1])}
dict_token3_sorted = {k: v for k, v in sorted(dict_token3.items(), key=lambda item: item[1])}
dict_token4_sorted = {k: v for k, v in sorted(dict_token4.items(), key=lambda item: item[1])}
dict_token5_sorted = {k: v for k, v in sorted(dict_token5.items(), key=lambda item: item[1])}
```

```python
def find_common(n):
  
  top20_token1 = list(dict_token1_sorted.keys())[-n:]
  top20_token2 = list(dict_token2_sorted.keys())[-n:]
  top20_token3 = list(dict_token3_sorted.keys())[-n:]
  top20_token4 = list(dict_token4_sorted.keys())[-n:]
  top20_token5 = list(dict_token5_sorted.keys())[-n:]
  
  common = []
  for i in top20_token1:
    if(i in top20_token2 and i in top20_token3 and i in top20_token4 and i in top20_token5):
      common.append(i)
  return common
```
